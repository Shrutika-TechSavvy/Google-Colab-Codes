```python
%%writefile parallel_sum.c
#include <stdio.h>
#include <omp.h>
#include <stdlib.h>
#include <time.h>

int main() {
    int n, i;
    double start_time, end_time;
    long long sum = 0;

    // Take dynamic input for array size
    printf("Enter number of elements: ");
    scanf("%d", &n);

    //  Dynamically allocate memory
    int *arr = (int*)malloc(n * sizeof(int));

    // Input elements from user
    printf("Enter %d elements:\n", n);
    for(i = 0; i < n; i++) {
        scanf("%d", &arr[i]);
    }

    //  Print CPU details
    int num_cores = omp_get_num_procs();
    printf("\nNumber of CPU Cores (Processors): %d\n", num_cores);

    // Start measuring execution time
    start_time = omp_get_wtime();

    //  Parallel region for sum using reduction
    #pragma omp parallel for reduction(+:sum)
    for(i = 0; i < n; i++) {

        //  Display number of threads only once
        if(i == 0) {
            printf("Number of Threads Used: %d\n", omp_get_num_threads());
        }

        sum += arr[i];
    }

    // End time
    end_time = omp_get_wtime();

    //  Print output results
    printf("\nSum of array elements = %lld\n", sum);
    printf("Execution Time: %f seconds\n", end_time - start_time);

    free(arr);
    return 0;
}

```
