```python
%%cuda
#include <stdio.h>

__global__ void hello_world() {
    printf("Hello from GPU!\n");
}

int main() {
    hello_world<<<1, 1>>>();
    cudaDeviceSynchronize();
    return 0;
}
```

```python
# Write the CUDA code to a file
with open('hello_world.cu', 'w') as f:
    f.write("""
#include <stdio.h>

__global__ void hello_world() {
    printf("Hello from GPU!\\n");
}

int main() {
    hello_world<<<1, 1>>>();
    cudaDeviceSynchronize();
    return 0;
}
""")
```

```python
# Compile the CUDA code using nvcc
!nvcc hello_world.cu -o hello_world
```

```python
# Run the compiled program
!./hello_world
```

```python
%%writefile add_two_numbers.cu
#include <stdio.h>

__global__ void add(int a, int b, int *c) {
    *c = a + b;
}

int main() {
    int a = 5;
    int b = 7;
    int c = 0;
    int *dev_c;

    cudaMalloc((void **)&dev_c, sizeof(int));

    add<<<1, 1>>>(a, b, dev_c);
    cudaMemcpy(&c, dev_c, sizeof(int), cudaMemcpyDeviceToHost);

    printf("Result: %d + %d = %d\n", a, b, c);

    cudaFree(dev_c);
    return 0;
}
```

```python
!nvcc add_two_numbers.cu -o add_two_numbers.cu
!./add_two_numbers.cu
```
