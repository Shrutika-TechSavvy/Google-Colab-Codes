```python
%%writefile max_openmp.c
#include <stdio.h>
#include <omp.h>

int main() {
    int arr[] = {3, 7, 2, 9, 5, 12, 1, 8};
    int n = 8;
    int max_val = arr[0];

    #pragma omp parallel for
    for(int i = 1; i < n; i++) {
        if(arr[i] > max_val) {
            max_val = arr[i];
        }
    }

    printf("Maximum element is %d\n", max_val);
    return 0;
}
```

```python
%%writefile max_openmp1.c
#include <stdio.h>
#include <omp.h>
#include <stdlib.h>

int main() {
    int n;
    printf("Enter number of elements: ");
    scanf("%d", &n);

    int *arr = (int *)malloc(n * sizeof(int));
    if (!arr) {
        printf("Memory allocation failed!\n");
        return 1;
    }

    printf("Enter %d elements: ", n);
    for(int i = 0; i < n; i++) {
        scanf("%d", &arr[i]);
    }

    int max_val = arr[0];
    int used_threads = 0;

    double start = omp_get_wtime(); // start time

    #pragma omp parallel for reduction(max:max_val)
    for(int i = 1; i < n; i++) {
        if (omp_get_thread_num() == 0) {
            used_threads = omp_get_num_threads();
        }
        if(arr[i] > max_val) {
            max_val = arr[i];
        }
    }

    double end = omp_get_wtime(); // end time

    printf("\nMaximum element is: %d\n", max_val);
    printf("Execution Time: %f seconds\n", end - start);
    printf("Total CPU cores available: %d\n", omp_get_num_procs());
    printf("Threads used by OpenMP: %d\n", used_threads);

    free(arr);
    return 0;
}
```
