```python
%%writefile max_openmp.c
#include <stdio.h>
#include <omp.h>

int main() {
    int arr[] = {3, 7, 2, 9, 5, 12, 1, 8};
    int n = 8;
    int max_val = arr[0];

    #pragma omp parallel for reduction(max:max_val)
    for(int i = 1; i < n; i++) {
        if(arr[i] > max_val) {
            max_val = arr[i];
        }
    }

    printf("Maximum element is %d\n", max_val);
    return 0;
}
```

```python
!gcc -fopenmp max_openmp.c -o max_openmp
```
